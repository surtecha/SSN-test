```python
#Getting the Monthly Smoothed Sunspot Data from Daily Data

sun_month=pd.read_csv("C:/Users/SSA/Desktop/Jesto Research/Original SUNSPOTS Datasets for Jupyter/SN_m_tot_V2.0.csv")
sun_month["year_month"]=sun_month["Year"].astype(str) +"-"+sun_month["Month"].astype(str)
sun_month=sun_month.drop(["date_fraction","sd","no_obser","provisional"],axis=1)
sun_month=sun_month[12:-5]
yhat_ = signal.savgol_filter(sun_month["sunspots"], 13, 1)

sun_month["Smoothed Savgol"]=yhat_
sun_month
```
